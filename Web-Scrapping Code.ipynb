```python
# City = /^(.*?), 

#, (\w+\s?\w+) --> State Name

# Public/private = \b(Public\/Govt|Private)\b

# ratings = \d+\.\d+\/5

# reviews = \/5(\d+)
```

```python
college_name = []
rating=[]
review=[]
NIRF_Ranking = []
City = []
State = []
Ownership =[]
Fees = []
BTech_Courses = [] 
MTech_Courses = []

for i in range(1,12):
    url=f"https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india?page={i}&sort_by=3&stream=1"
    response=requests.get(url)
    soup = BeautifulSoup(response.text)
    for i in soup.find_all("div",class_="tupple_right_block d-none d-md-block"):

        # print(i.find('h3',class_='college_name').text)
        if i.find('h3',class_='college_name').text:
            college_name.append(i.find('h3',class_='college_name').text)
        else:
            college_name.append(np.nan)
            
    if soup.find_all("div",class_="ranking_strip d-md-none"):
        for i in soup.find_all("div",class_="ranking_strip d-md-none"):
        # print(i.text)
            if re.findall(r"NIRF Ranking:(\s*\d+(-\d+)?)",i.text):
                NIRF_Ranking.append("".join(re.findall(r"NIRF Ranking:(\s*\d+(-\d+)?)",i.text)[0][0]))
            else:
                NIRF_Ranking.append(np.nan)
    else:
        NIRF_Ranking.append(np.nan)

    
    for i in soup.find_all("div",class_ = "content_block d-none d-md-block d-md-flex flex-row justify-content-between"):
          # print(i.text)
        
         if re.findall("^(.*?),",i.text):
           City.append("".join(re.findall("^(.*?),",i.text)))
         else: 
           City.append(np.nan)

         if re.findall(", (\w+\s?\w+)",i.text):
             State.append("".join(re.findall(", (\w+\s?\w+)",i.text)))
         else: 
             State.append(np.nan)

         if re.findall(r"\b(Public\/Govt|Private)\b", i.text):
             Ownership.append("".join(re.findall(r"\b(Public\/Govt|Private)\b", i.text)))
         else:
             Ownership.append(np.nan)

         if re.findall("\d+\.\d+",i.text):
             rating.append("".join(re.findall("\d+\.\d+",i.text)))
         else: 
             rating.append(np.nan)

        
         if re.findall("\/5(\d+)",i.text):
             review.append("".join(re.findall("\/5(\d+)",i.text)))
         else: 
             review.append(np.nan)

    for i in soup.find_all("div",class_ = "snippet_block"):
         # print(i.text)
        
        
        if re.findall("Fees\s*:\s*([₹\d.]+[A-Za-z]*\sLakhs|[₹\d.]+[A-Za-z]*\sK)",i.text):
            Fees.append("".join(re.findall("Fees\s*:\s*([₹\d.]+[A-Za-z]*\sLakhs|[₹\d.]+[A-Za-z]*\sK)",i.text)))
        else: 
            Fees.append(np.nan)

        if re.findall("B\.E\s\/B\.Tech\s\((\d+\s)Courses",i.text):
             BTech_Courses.append("".join(re.findall("B\.E\s\/B\.Tech\s\((\d+\s)Courses",i.text)))
        else: 
             BTech_Courses.append(np.nan)

        if re.findall("M\.E\s\/M\.Tech.\s\((\d+\s)Courses",i.text):
            MTech_Courses.append("".join(re.findall("M\.E\s\/M\.Tech.\s\((\d+\s)Courses",i.text)))
        else: 
            MTech_Courses.append(np.nan)
```

```python
print(len(college_name))
print(len(rating))
print(len(review))
print(len(NIRF_Ranking))
print(len(City))
print(len(State))
print(len(Ownership))
print(len(Fees))
print(len(BTech_Courses))
print(len(MTech_Courses))
```
